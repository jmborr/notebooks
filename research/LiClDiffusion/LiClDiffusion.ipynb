(<a href='#Table of Contents'>Top</a>)<a id='goal'></a><h3>Goal</h3>
The intent is to show that concentrated LiCl in water Leads to a model of continuous diffusion for water. The hypothesis is that LiCl disrupts the hydrogen network such that water behaves more like a Lennard-Jones liquid.

(<a href='#Experiment'>Top</a>)<a id='Experiment.details'></a><h3>Details of the experiment</h3>

QENS data for aqeous LiCl, with six water molecules per LiCl pair.  
Collected dynamic ranges was [-100,100]ueV  and Q-range [0.3,1.9] for data collected with the Si(111) analyzer (3.4 ueV FWHM nominal resolution) and [-660, 660]ueV and Q-range [0.5,3.7] for data collected with the Si(311) analyzer (15.0 ueV FWHM nominal resolution).

For the $D\cdot Q^2$ plots, Eugene used Si(111) data for Q=0.3,0.5,0.7,0.9 A^-1, and Si(311) data for the rest of Q values (1.1,1.3,1.5,1.7,1.9)A^-1. All fall onto the same $D\cdot Q^2$ line, from the smallest to the largest Q.

(<a href='#Experiment'>Top</a>)<a id='Experiment.stretchedfit'></a><h3>Fitting experiment with Stretched Exponential</h3>
We store the Mantid session for fitting of data at 240K in subdirectory <i>experiment/stretchedExFTfits</i>. We used the following model:  
$Resolution \otimes (EISF \delta(E) + (1-EISF) StretchedExFT(E)) + LinearBackground$  
and fitted data from the Si111 and Si311 analyzers.  
We calculate $FWHM(Q) = \hbar / \bar{\tau}(Q)$, where $\bar{\tau}(Q) = \tau(Q)/\beta(Q) Gamma(1/\beta(Q))$  
<center><a href="files/experiment/stretchedExFTfits/FWHM.jpeg"><img src="files/experiment/stretchedExFTfits/FWHM.jpeg" width="300" height="300" alt="experiment/stretchedExFTfits/FWHM.jpeg"></a>  <i>experiment/stretchedExFTfits/FWHM.jpeg</i></center>  
Observations from the plot:

* $FWHM(Q)$ from Si111 and Si311 reflections overlap in the common range of Q-values
* $FWHM(Q) \propto Q^2$ if using only Si111 data. There's deviation if looking at higher Q-values (the Si311 data).

(<a href='#Experiment'>Top</a>)<a id='Experiment.stretchedfitAllT'></a><h3>Fitting experiment with Stretched exponential at all temperatures</h3>
Fixes Github issue #[7](https://github.com/jmborr/LiClDiffusion/issues/7).

Under <i>experiment/</i>, I created directories <i>stretchedExFTfits240K</i>,...<i>stretchedExFTfits300K</i> containing Mantid sessions. Each session is a fitting of the Si111 and Si311 data to a stretched exponential.  

From these sessions I extracted the Q-dependence of the average relaxation time, computed from the stretched exponential fit. In directory <i>experiment/analysis/</i> I stored files <code>params_240K_Si*_versusQ_avg_tau.dat params_260K_Si*_versusQ_avg_tau.dat, params_280K_Si*_versusQ_avg_tau.dat, params_300K_Si*_versusQ_avg_tau.dat</code>.  

For each temperature and using the both Si111 and Si311 reflections, I created a composite Q-dependence of the average relaxation file, which I store as files <code>params_240K_versusQ_avg_tau.dat, params_260K_versusQ_avg_tau.dat, params_280K_versusQ_avg_tau.dat, params_300K_versusQ_avg_tau.dat</code>.  

I created <code>xmgrace</code> figures plotting this Q-dependence in log-log mode in files <code>params_240K_versusQ_avg_tau.agr, params_260K_versusQ_avg_tau.agr, params_280K_versusQ_avg_tau.agr, params_300K_versusQ_avg_tau.agr</code>.

From the previous <code>xmgrace</code> files, I created the $Q^2$ dependence of $FWHM=0.658/<\tau>$ and save them as files <code>params_240K_versusQ2_FWHM.agr, params_260K_versusQ2_FWHM.agr, params_280K_versusQ2_FWHM.agr, params_300K_versusQ2_FWHM.agr</code>.

I then fitted the $Q^2$ dependence of $FWHM$ with the $FWHM$ expression for the translational component of the Teixeira-model, obtaining a diffusion coefficient ($D$) and a waiting time ($\tau_0$) for each temperature. These parameters I saved them in xmgrace figure <code>D_tau_Teixeira_vs_LiCl.agr</code>. This figure also contains the temperature dependence of the $D$ and  $\tau_0$ for pure water, extracted from the Teixeira85 paper.

(<a href='#TIP3P.LiClsimulations'>Top</a>)<a id='TIP3P.LiClsimulations.initialSystem'></a><h3>Creation of the initial system</h3>
The lowest value of the momentum transfer (Q) in the QENS experiments is 0.3 inverse Angstroms. Thus, we need a simulation box of dimensions bigger then $2\pi/0.3=21$Angstroms. We deem that a 30Angstroms simulation box is more than enough.

<b>Guessing the number of water molecules</b>  
We guess the number of water molecules in a 30A box. In <i>preparing/watBox30/guessNwat</i> command  
<code>tleap11 -f in.tleap > in.log</code>  
generates a pure water box of 30Å size with 1128 water molecules (files watBox30.crd, watBox30.pdb, and waterBox30.top).
<a href="files/preparing/watBox30/guessNwat/watBox30.png"><img src="files/preparing/watBox30/guessNwat/watBox30.png" width="200" height="200" alt="preparing/watBox30/guessNwat/watBox30.png"></a>  <i>preparing/watBox30/guessNwat/watBox30.png</i>

<b>Packing the molecules</b>  
At experimental concentrations (six water molecules per LiCl pair) we need 188 LiCl pairs if we are to use 1128 water molecules. In <i>preparing/watBox30/packmol/</i> we create a system of 1128 water molecules along with 188 Cl- and 188 Li+ ions.  
<code>packmol < packmol.inp</code>  
generates file <i>system30.0.pdb</i>  
<a href="files/preparing/watBox30/packmol/system30.0.png"><img src="files/preparing/watBox30/packmol/system30.0.png" width="200" height="200" alt="preparing/watBox30/packmol/system30.0.png"></a>  <i>preparing/watBox30/packmol/system30.0.png</i>  
We reshuffle the ordering of the atoms in system30.0.pdb so that Li+ shows first, followed by Cl-.  
We insert the follwing renaming:

* Li+ as LIT
* Cl- as CLA
* "O   WAT " as "OH2 TIP3"
* "WAT " as "TIP3"  
We store the new file as <i>system30.pdb</i>. Then we correct the numering due to the previous reshuffling by loading system30.pdb in VMD, then saving the coordinates as <i>system30_vmd.pdb</i>.

<b>Creating the PSF file</b>  
In <i>preparing/watBox30/relax/</i> we run <code>vmd system30_vmd.pdb</code>, then Extensions → Modelling → Automatic PSF Builder. We:

* delete the default topology files and load top_all27_prot_lipid.inp, then press “Load input files”.
* press “Guess and split chains”. Three chains will be created, one for Lithium, one for Chorine, one for water.
* press “Create chains” and subsequent pop ups.  
Files <i>system30_vmd_autopsf.psf</i> and <i>system30_vmd_autopsf.pdb</i> are generated.

<b>Relaxing the initial conformation</b>  
In <i>preparing/watBox30/relax/</i> we run <code>namd2 +idlepoll +p32 minimize.conf> minimize.log</code> in camm2 runs an energy minimization of the conformation followed by a short run in the NPT ensemble with T=290K and P=1atm (timestep=0.1fs). We set the cell size to 36Angstroms because the atoms are too crammed. Recall that the 30A box was appropriate for water density, without the presence of the LiCl.  
I had to select high values for parameters cutoff and pairlistdist, as well as a small value for stepspercycl (see file minimize.conf). All this because at the beginning of the short run atoms are moving too fast.  

We extract thermodynamic quantities from file <i>minimize.log</i> into <i>minimize.dat</i> for the minimization step and <i>minimize_2.dat</i> for the subsequent heating to 290K

We compare the diffusion coefficient we obtain from the fit, $D=0.055meV\cdot A^2/\hbar$ (see figure above) with the diffusion coefficient we obtained from the <a href='#WaterSimulations.diffusion'>atomic displacements</a> ($D=1.02\cdot 10^{-5}cm^2/s$):

Using the random jump-diffusion model of , we have $FWHM(Q)=\frac{DQ^2}{1+DQ^2\tau_0}$, with $\tau_0$ being the time spent in between jumps, and the mean-square jump length $<l^2> = 6D\tau_0$. We can fit all the points to the $FWHM$ expression, obtaining $D=0.606meV\cdot A^2/\hbar$, $\tau_0=0.4956meV/\hbar=0.75ps$, and $<l^2>=(0.42A)^2$.

By attending at the Q-dependence of FWHM(Q) <i>at high Q</i> we can see that:

* In the stretched exponential fit,  $FWHM \approx Q^\nu$ with $\nu>2$.

(<a href='#TIP4P2005.LiClsimulations'>Top</a>)<a id='TIP4P2005.LiClsimulations.betaexponent'></a><h3>Comparison of stretching exponent Beta</h3>

We fit the simulated $I(Q,t)$ to a model of a stretched exponential plus a flat background. The plot below shows the Q-dependence of the stretching exponent for the following systems:

* LiCl at 240K with no removal of whole center of mass (CM) drift.
* LiCl at 240K with removal of CM drift (<code>rms2first</code> trajectory). Fits stored in Mantid session <i>TIP4P2005/LiCl/240K/production/fitStretchExp</i>.
* water at 240K with no removal of CM drift.
* water at 240K with removal of CM drift (<code>rms2first</code> trajectory). Fits stored in Mantid session <i>TIP4P2005/water/240K/production/fitStretchExp</i>.
* experiment at 240K.

<center><a href="files/TIP4P2005/analysis/betaExponent_rms2first.png"><img src="files/TIP4P2005/analysis/betaExponent_rms2first.png" width="400" height="400" alt="TIP4P2005/analysis/betaExponent_rms2first.png"></a>  <i>TIP4P2005/analysis/betaExponent_rms2first.png</i></center>

$\beta(Q)$ is very similar for the LiCl and water systems if we not remove the artificial CM drift because this drift determines the decay of $I(Q,t)$. The stretching exponents are very different once we remove the drift, underlying the different dynamics of the two systems.  
Water has a exponent relatively close to 1, thus it can be modeled with a single Lorentzian as expected of water at low temperatures. At higher temperatures the exponent decays faster with $Q$ due to the rotational diffusion:

(<a href='#meetings'>Top</a>)<a id='meetings.20160812'></a><h3>Eugene told me the basics of the project</h3>
Eugene performed QENS measurements on a concentrated solution of LiCl and found out that a model of "continous diffussion" fits the data. He used a stretched exponential and found that $\bar{\tau}=\frac{\tau}{\beta}\Gamma(\frac{1}{\beta})\propto Q^2$ single lorentzian and found $FWHM $. Does pure water follow the same mechanishm?

In the Teixeira85.pdf paper, authors fitted QENS of pure water with a single Lorentzian, and plotted it versus $Q^2$. At high Q ($Q>1.0$) we see a deviation from linearity even at room temperatures, due to the jump-diffusion mechanism:

$\Gamma = \hbar \frac{DQ^2}{1+DQ^2\tau_0}$  and $D=\frac{<l^2>}{6\tau_0}$

<a href="files/analysis/teixeira85.png"><img src="files/analysis/teixeira85.png" width="400" height="400" alt="analysis/teixeira85.png"></a>  <i>analysis/teixeira85.png</i>

A plot of $\tau_0$ versus temperature, shows sharp increase at low T, with $\tau_0=22.7 ps$ at $T=253K$: 

<a href="files/analysis/teixeira_tao0.png"><img src="files/analysis/teixeira_tao0.png" width="400" height="400" alt="analysis/teixeira_tao0.png"></a>  <i>analysis/teixeira_tao0.png</i>

A plot of $\hbar/\tau_1$ versus temperature, with $\tau_1$ an associated rotational diffusion relaxation time shows $\hbar/\tau_1 \simeq 0.3 meV$  at 240K. This is roughly the FWHM of the fast lorentzian in the two lorentzian fit.
